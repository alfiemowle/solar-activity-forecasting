# LSTM Forecasting with Cross-Validation on Sunspot Data

This notebook trains and evaluates an LSTM model for forecasting smoothed sunspot area from historical data.

Key features:
- Time-series cross-validation using `TimeSeriesSplit`
- Noise-augmented training sequences
- Gradient clipping, learning-rate scheduling, early stopping
- Aggregated cross-validation metrics (MAE, RMSE, R²)
- Final model trained on all seen data and evaluated on a held-out test set


```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import os
import random

# -------------------------
# 1. Set Random Seeds for Reproducibility
# -------------------------
def set_seed(seed=42):
    """
    Sets the seed for generating random numbers to ensure reproducibility.
    """
    torch.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed()

# -------------------------
# 2. Device Configuration
# -------------------------
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# -------------------------
# 3. Load and Prepare the Dataset
# -------------------------

# Load the dataset
df = df_smooth.copy()

# Ensure 'Date' is in datetime format
df_smooth['Date'] = pd.to_datetime(df_smooth['Date'])

# Use the 'Smoothed_Total' column as the sequence data
total_series = df_smooth['Smoothed_Total'].values

# -------------------------
# 4. Split Data into Seen (Training + Validation) and Unseen (Test) Sets
# -------------------------
split_seen_unseen = int(len(total_series) * 0.8)  # 80% for training + validation, 20% for testing
seen_data = total_series[:split_seen_unseen]
unseen_data = total_series[split_seen_unseen:]

print(f"Total data length: {len(total_series)}")
print(f"Seen data length: {len(seen_data)}")
print(f"Unseen data length: {len(unseen_data)}")

# -------------------------
# 5. Create Sequences from Seen Data
# -------------------------
seq_length = 300  

def create_sequences(data, seq_length):
    """
    Creates input-output sequences from the time series data.
    
    Args:
        data (numpy.ndarray): The time series data.
        seq_length (int): The length of each input sequence.
    
    Returns:
        Tuple of numpy.ndarrays: (sequences, targets)
    """
    xs = []
    ys = []
    for i in range(len(data) - seq_length):
        x = data[i:(i + seq_length)]
        y = data[i + seq_length]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

X, y = create_sequences(seen_data, seq_length)
print(f"Total Sequences in Seen Data: {len(X)}")

# -------------------------
# 6. Define the PyTorch Dataset Class
# -------------------------
class SequenceDataset(Dataset):
    def __init__(self, sequences, targets, add_noise=False, noise_std=0.01):
        """
        Initializes the dataset with sequences and targets.
        
        Args:
            sequences (numpy.ndarray): Input sequences.
            targets (numpy.ndarray): Corresponding target values.
            add_noise (bool): Flag to add noise to the input sequences.
            noise_std (float): Standard deviation of the Gaussian noise.
        """
        self.sequences = sequences
        self.targets = targets
        self.add_noise = add_noise
        self.noise_std = noise_std

    def __len__(self):
        return len(self.sequences)

    def __getitem__(self, idx):
        sequence = self.sequences[idx]
        target = self.targets[idx]
        sequence = sequence.reshape(-1, 1)  # Reshape for LSTM input

        if self.add_noise:
            # Generate Gaussian noise
            noise = np.random.normal(0, self.noise_std, sequence.shape)
            sequence = sequence + noise
            # Ensure the sequence remains within [0, 1] after noise addition
            sequence = np.clip(sequence, 0, 1)

        return torch.FloatTensor(sequence), torch.FloatTensor([target])

# -------------------------
# 7. Define the LSTM Forecasting Model
# -------------------------
class LSTMForecast(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, output_dim, dropout=0.3):
        """
        Initializes the LSTM forecasting model.
        
        Args:
            input_dim (int): Number of input features.
            hidden_dim (int): Number of features in the hidden state.
            num_layers (int): Number of recurrent layers.
            output_dim (int): Number of output features.
            dropout (float): Dropout probability.
        """
        super(LSTMForecast, self).__init__()
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers

        # LSTM layer with dropout
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True, dropout=dropout)

        # Batch Normalization
        self.bn = nn.BatchNorm1d(hidden_dim)

        # Fully connected layer
        self.fc = nn.Linear(hidden_dim, output_dim)

        # Sigmoid activation to constrain outputs between 0 and 1
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # Initialize hidden and cell states with zeros
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim).to(x.device)

        # Forward propagate LSTM
        out, _ = self.lstm(x, (h0, c0))  # out: tensor of shape (batch_size, seq_length, hidden_dim)

        # Apply Batch Normalization on the last LSTM output
        out = out[:, -1, :]  # Take the output of the last time step
        out = self.bn(out)

        # Decode the hidden state of the last time step
        out = self.fc(out)

        # Apply Sigmoid activation
        out = self.sigmoid(out)

        return out

# -------------------------
# 8. Initialize Scalers and Scale the Data
# -------------------------
# Initialize separate scalers for inputs (X) and targets (y)
scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))

# Fit scalers on the entire seen data
scaler_X.fit(X.reshape(-1, 1))
scaler_y.fit(y.reshape(-1, 1))

# Scale the entire dataset
X_scaled = scaler_X.transform(X.reshape(-1, 1)).reshape(X.shape)
y_scaled = scaler_y.transform(y.reshape(-1, 1)).flatten()

# -------------------------
# 9. Set Up Time Series Cross-Validation
# -------------------------
n_splits = 5  # Number of folds
tscv = TimeSeriesSplit(n_splits=n_splits)

# Model Training Parameters
input_dim = 1
hidden_dim = 64
num_layers = 6
output_dim = 1
dropout = 0.2
batch_size = 16
num_epochs = 200
patience = 10
noise_std = 0.001
learning_rate = 0.001

# Store performance metrics for each fold
fold_metrics = []

# -------------------------
# 10. Cross-Validation Loop
# -------------------------
for fold, (train_index, val_index) in enumerate(tscv.split(X_scaled)):
    print(f"\n=== Fold {fold + 1} ===")

    # Split data based on indices
    X_train_fold, X_val_fold = X_scaled[train_index], X_scaled[val_index]
    y_train_fold, y_val_fold = y_scaled[train_index], y_scaled[val_index]

    # Create Datasets
    train_dataset = SequenceDataset(X_train_fold, y_train_fold, add_noise=True, noise_std=noise_std)
    val_dataset = SequenceDataset(X_val_fold, y_val_fold, add_noise=False)

    # Create DataLoaders
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

    # Initialize the model
    model = LSTMForecast(input_dim, hidden_dim, num_layers, output_dim, dropout).to(device)

    # Define loss function and optimizer
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

    # Define learning rate scheduler
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5, verbose=False)

    # Early stopping parameters
    best_val_loss = float('inf')
    patience_counter = 0

    # Lists to store losses
    train_losses = []
    val_losses = []

    # Directory to save the best model for this fold
    model_dir = f'models/fold_{fold + 1}'
    os.makedirs(model_dir, exist_ok=True)
    best_model_path = os.path.join(model_dir, 'best_model.pth')

    # Training loop
    for epoch in range(num_epochs):
        model.train()
        train_loss = 0

        for sequences, targets in train_loader:
            sequences = sequences.to(device)
            targets = targets.to(device)
            optimizer.zero_grad()
            outputs = model(sequences)
            loss = criterion(outputs, targets)
            loss.backward()

            # Gradient Clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            optimizer.step()
            train_loss += loss.item()

        train_loss /= len(train_loader)
        train_losses.append(train_loss)

        # Validation phase
        model.eval()
        val_loss = 0
        with torch.no_grad():
            for sequences, targets in val_loader:
                sequences = sequences.to(device)
                targets = targets.to(device)
                outputs = model(sequences)
                loss = criterion(outputs, targets)
                val_loss += loss.item()
        val_loss /= len(val_loader)
        val_losses.append(val_loss)

        # Step the scheduler
        scheduler.step(val_loss)
        current_lr = scheduler.get_last_lr()

        print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {train_loss:.6f}, Validation Loss: {val_loss:.6f}, Current Learning Rate: {current_lr}')

        # Early stopping logic based on validation loss
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            patience_counter = 0
            torch.save(model.state_dict(), best_model_path)
        else:
            patience_counter += 1

        if patience_counter >= patience:
            print(f"Early stopping triggered after {epoch+1} epochs.")
            break

    # Load the best model
    model.load_state_dict(torch.load(best_model_path))

    # Evaluate on validation set
    model.eval()
    predictions = []
    actuals = []

    with torch.no_grad():
        for sequences, targets in val_loader:
            sequences = sequences.to(device)
            outputs = model(sequences)
            predictions.append(outputs.cpu().numpy())
            actuals.append(targets.cpu().numpy())

    # Concatenate all predictions and actuals
    predictions = np.concatenate(predictions).flatten()
    actuals = np.concatenate(actuals).flatten()

    # Inverse transform predictions and actuals
    predictions_inv = scaler_y.inverse_transform(predictions.reshape(-1, 1)).flatten()
    actuals_inv = scaler_y.inverse_transform(actuals.reshape(-1, 1)).flatten()

    # Calculate metrics
    mae = mean_absolute_error(actuals_inv, predictions_inv)
    rmse = np.sqrt(mean_squared_error(actuals_inv, predictions_inv))
    r2 = r2_score(actuals_inv, predictions_inv)

    print(f"Fold {fold + 1} Metrics:")
    print(f"MAE: {mae:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"R² Score: {r2:.4f}")

    # Store the metrics
    fold_metrics.append({
        'fold': fold + 1,
        'MAE': mae,
        'RMSE': rmse,
        'R2_Score': r2
    })

# -------------------------
# 11. Aggregate and Display Cross-Validation Metrics
# -------------------------
metrics_df = pd.DataFrame(fold_metrics)
print("\n=== Cross-Validation Metrics ===")
print(metrics_df)

# Calculate mean and standard deviation of metrics
mean_mae = metrics_df['MAE'].mean()
std_mae = metrics_df['MAE'].std()

mean_rmse = metrics_df['RMSE'].mean()
std_rmse = metrics_df['RMSE'].std()

mean_r2 = metrics_df['R2_Score'].mean()
std_r2 = metrics_df['R2_Score'].std()

print("\n=== Cross-Validation Performance ===")
print(f"Average MAE: {mean_mae:.2f} ± {std_mae:.2f}")
print(f"Average RMSE: {mean_rmse:.2f} ± {std_rmse:.2f}")
print(f"Average R² Score: {mean_r2:.4f} ± {std_r2:.4f}")

# -------------------------
# 12. Final Model Training on Entire Training Data
# -------------------------
print("\n=== Training Final Model on Entire Training Data ===")

# Combine all training data
X_train_all = X_scaled
y_train_all = y_scaled

# Create Dataset and DataLoader
train_dataset_all = SequenceDataset(X_train_all, y_train_all, add_noise=True, noise_std=noise_std)
train_loader_all = DataLoader(train_dataset_all, batch_size=batch_size, shuffle=True)

# Initialize the final model
model_final = LSTMForecast(input_dim, hidden_dim, num_layers, output_dim, dropout).to(device)

# Define loss function and optimizer
criterion_final = nn.MSELoss()
optimizer_final = torch.optim.Adam(model_final.parameters(), lr=learning_rate)

# Define learning rate scheduler
scheduler_final = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer_final, mode='min', 
                                                             factor=0.5, patience=5, verbose=False)

# Early stopping parameters
best_val_loss_final = float('inf')
patience_counter_final = 0

# Lists to store losses
train_losses_final = []
val_losses_final = []

# Directory to save the best final model
model_dir_final = 'models/final_model'
os.makedirs(model_dir_final, exist_ok=True)
best_model_path_final = os.path.join(model_dir_final, 'best_model.pth')

# Training loop for final model
for epoch in range(num_epochs):
    model_final.train()
    train_loss = 0

    for sequences, targets in train_loader_all:
        sequences = sequences.to(device)
        targets = targets.to(device)
        optimizer_final.zero_grad()
        outputs = model_final(sequences)
        loss = criterion_final(outputs, targets)
        loss.backward()

        # Gradient Clipping
        torch.nn.utils.clip_grad_norm_(model_final.parameters(), max_norm=1.0)

        optimizer_final.step()
        train_loss += loss.item()

    train_loss /= len(train_loader_all)
    train_losses_final.append(train_loss)

    # Since we don't have a separate validation set, use training loss for scheduler
    scheduler_final.step(train_loss)
    current_lr_final = scheduler_final.get_last_lr()

    print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {train_loss:.6f}, Current Learning Rate: {current_lr_final}')

    # Early stopping logic based on training loss plateau
    if train_loss < best_val_loss_final:
        best_val_loss_final = train_loss
        patience_counter_final = 0
        torch.save(model_final.state_dict(), best_model_path_final)
    else:
        patience_counter_final += 1

    if patience_counter_final >= patience:
        print(f"Early stopping triggered after {epoch+1} epochs.")
        break

# Load the best final model
model_final.load_state_dict(torch.load(best_model_path_final))

# -------------------------
# 13. Prepare and Scale the Test Set
# -------------------------
print("\n=== Preparing Test Set ===")

# Create sequences from unseen (test) data
X_test, y_test = create_sequences(unseen_data, seq_length)
print(f"Test Sequences: {len(X_test)}")  # Expected: len(unseen_data) - seq_length

# Scale the test data using the same scalers fitted on training data
X_test_scaled = scaler_X.transform(X_test.reshape(-1, 1)).reshape(X_test.shape)
y_test_scaled = scaler_y.transform(y_test.reshape(-1, 1)).flatten()

print("\n=== Test Scaled Range ===")
print(f"Test Scaled Range (X): {X_test_scaled.min()} - {X_test_scaled.max()}")
print(f"Test Scaled Range (y): {y_test_scaled.min()} - {y_test_scaled.max()}")

# Create the Test Dataset without adding noise
test_dataset = SequenceDataset(X_test_scaled, y_test_scaled, add_noise=False)

# Create the Test DataLoader
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)

# -------------------------
# 14. Evaluate the Final Model on the Test Set
# -------------------------
print("\n=== Evaluating Final Model on Test Set ===")

model_final.eval()
predictions_test = []
actuals_test = []

with torch.no_grad():
    for sequences, targets in test_loader:
        sequences = sequences.to(device)
        outputs = model_final(sequences)
        predictions_test.append(outputs.cpu().numpy())
        actuals_test.append(targets.cpu().numpy())

# Concatenate predictions and actuals
predictions_test = np.concatenate(predictions_test).flatten()
actuals_test = np.concatenate(actuals_test).flatten()

# Inverse transform predictions and actuals
predictions_test_inv = scaler_y.inverse_transform(predictions_test.reshape(-1, 1)).flatten()
actuals_test_inv = scaler_y.inverse_transform(actuals_test.reshape(-1, 1)).flatten()

# Create Test Dates for Plotting (Assuming 'Date' is aligned with 'total_series')
# Adjust 'test_start' based on how sequences were created
# Here, 'split_seen_unseen' is the index where unseen_data starts
test_start = split_seen_unseen + seq_length
test_end = test_start + len(y_test)
test_dates = df_smooth['Date'].values[test_start:test_end]

print("\n=== Test Data Evaluation ===")
print(f"Length of test_dates: {len(test_dates)}")        # Expected: len(y_test)
print(f"Length of actuals: {len(actuals_test_inv)}")      # Expected: len(y_test)
print(f"Test Data Range: {actuals_test_inv.min()} - {actuals_test_inv.max()}")
print("First few test_dates:", test_dates[:10])
print("First few actuals:", actuals_test_inv[:10])
print("First few predictions:", predictions_test_inv[:10])

# Plot the actual vs predicted values on test data
plt.figure(figsize=(12, 6))
plt.plot(test_dates, actuals_test_inv, label='Actual Total Sunspot Area')
plt.plot(test_dates, predictions_test_inv, label='Predicted Total Sunspot Area')
plt.xlabel('Date')
plt.ylabel('Total Sunspot Area')
plt.title('Actual vs Predicted Total Sunspot Area for Test Data')
plt.legend()
plt.show()

# Calculate Performance Metrics
mae_test = mean_absolute_error(actuals_test_inv, predictions_test_inv)
rmse_test = np.sqrt(mean_squared_error(actuals_test_inv, predictions_test_inv))
r2_test = r2_score(actuals_test_inv, predictions_test_inv)

print("\n=== Test Data Metrics ===")
print(f"Test MAE: {mae_test:.2f}")
print(f"Test RMSE: {rmse_test:.2f}")
print(f"Test R² Score: {r2_test:.4f}")
```
